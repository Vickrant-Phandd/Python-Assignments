```python
#1.	Using sklearn import the wine dataset
import pandas as pd
import numpy as np
import sklearn
```

```python
from sklearn.datasets import load_wine
```

```python
data = load_wine()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df["clss"] = data.target
df
```

```python
# X is input variables and Y is output Variables
X=df.loc[:,"alcohol":"proline"]
Y=df.loc[:,["clss"]]
```

```python
df.shape
```

```python
X
```

```python
Y
```

```python
#2.	Split the data into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2) 
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
#3.	Train the model
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators=20, random_state=0, max_features= 3, verbose =1)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
classifier.get_params
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
#print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
